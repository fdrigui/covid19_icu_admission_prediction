# Ingestão do *'raw data'*
O *raw data* **não** está sendo baixado automaticamento do site do *Kaggle*.<br>
Essa decisão foi tomada de maneira a evitar que mudanças no *raw data* afetem esse projeto.<br>
Imagine que o time de *Data Intteligence* do *Hospital Sírio-Libanês* adicione ou remova algum feature, ou então adicionem mais pacientes no dataset. Isso poderia fazer com que algum script ou nootebook viesse a quebrar, além de possivelmente mudar o score dos modelos de ML.<br>
Para evitar essa situação, o dataset foi baixado do Kaggle dia *16 de Setembro de 2021*, e pode ser encontrado nesse projeto no seguinte caminho: ```/data/raw/Kaggle_Sirio_Libanes_ICU_Prediction.xlsx```<br><br>
**Raw data** é um termo amplamente utilizado na Ciência de Dados para se referir aos **dados não tratados**, que são adquiridos no começo de uma análise. Nesse projeto vou dar preferência ao termo em ingles.

## Importe das bibliotecas necessárias

````python
import pandas as pd #Ferramenta de Análise e Manipulação de dados
````

## Leitura do *'raw data'*
* Leitura do raw data desse repositório do gitdum
* **df** vai ser o nome do ```Pandas.DataFrame``` com os dados importados 

````python
# Leitura do raw data desse  projeto no Github
df = pd.read_excel('https://github.com/fdrigui/covid19_icu_admission_prediction/raw/main/data/raw/Kaggle_Sirio_Libanes_ICU_Prediction.xlsx' )

# Imprime na tela todas as colunas em vez de as 10 primeiras e 10 ultimas
pd.set_option('max_columns', df.shape[1])
df.head()
````

## Uma visão geral do DataFrame

* 002 colunas são do tipo: Object
